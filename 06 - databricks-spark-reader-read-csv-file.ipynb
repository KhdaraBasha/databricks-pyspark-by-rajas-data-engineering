To read a CSV file in Databricks, use the following code:

_**Syntax:**_

**Method 1** (using `.format` and `.option`):

python
df = spark.read.format(file_type) \
    .option("inferSchema", True) \
    .option("header", True) \
    .option("sep", ",") \
    .schema(schema_df) \
    .load(file_location)

- `file_type`: e.g., "csv"
- `inferSchema`: automatically infers column types
- `header`: uses the first row as column names
- `sep`: delimiter (default is ',')
- `schema_df`: specify schema if needed
- `file_location`: path to the CSV file

**Method 2** (using `.csv` directly):

python
df = spark.read.csv("/path/to/file.csv", header=True, inferSchema=True)
df.show()

- `header=True`: uses the first row as column names
- `inferSchema=True`: automatically infers column types

### Read Single CSV file

```python
df = spark.read.format('csv')\
    .option('header','true')\
    .option('inferSchema','true')\
    .load('/Volumes/pysaprk_demo/default/sample_data/baby_names_2007_2009.csv')

# display sample data
display(df)

# count of rows
display(df.count() )
```

### Create own Schema

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

schema = StructType([
    StructField('year', IntegerType(), True),
    StructField('name', StringType(), True),
    StructField('gender', StringType(), True),
    StructField('count', IntegerType(), True),
    StructField('country', StringType(), True)
])

df = spark.read.csv('/Volumes/pysaprk_demo/default/sample_data/baby_names_*.csv', header=True, schema=schema)

display(df)
```

```python
display(df.printSchema())
```

```python
schema_alternate= StructType([
    StructField('year', IntegerType(), True),
    StructField('name', StringType(), True),
    StructField('gender', StringType(), True),
    StructField('count', IntegerType(), True),
    StructField('state', StringType(), True),
    StructField('occupation', StringType(), True)
])

df = spark.read.csv('/Volumes/pysaprk_demo/default/sample_data/baby_names_*.csv', header=True, schema=schema_alternate) 

display(df)
display(df.printSchema())
```
