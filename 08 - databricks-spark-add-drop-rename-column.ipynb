### **_[Add Column](url)_**:
Use withColumn() to create a new column or modify an existing one.

df.withColumn("new_column_name", value)

### **_[Rename Column](url)_**:
Use withColumnRenamed() to change the column name.

df.withColumnRenamed("old_name", "new_name")

### **_[DROP Column](url)_**:
Use drop() to remove a column from the DataFrame.

df.drop("column name)

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
employee_data = [
(10,"Raj","Kumar","1999","100","M",2000),
(20,"Rahul","Rajan","2002","200","F",8000),
(30,"Raghav","Manish","2010","100",None,6000),
(40,"Raja","Singh","2004","100","F",7000),
(50,"Rama","Krish","2008","400","M",1000),
(60,"Rasul","Kutty","2014","500","M",5000),
(70,"Kumar","Chand","2004","600","M",5000)
]

columns = ["employee_id","first_name","last_name","doj","dept_id","gender","salary"]

emp_df = spark.createDataFrame(employee_data, columns)

display(emp_df)
```

### Add new column using constant literal

```python
emp_df = (
    emp_df.withColumn("bonus", lit(0.15))
    .withColumn("bonus_amount", col("salary") * col("bonus"))
    .withColumn("bonus_amount", round(col("bonus_amount"), 2))
    .withColumn("total_salary", col("salary") + col("bonus_amount"))
    .withColumn("doj", to_date(col("doj"), "yyyy"))
)

emp_df.show()
```

### Add new column by calculation

```python
emp_df = (
    emp_df.withColumn("name", concat(col("first_name"), lit(", "), col("last_name")))
)

emp_df.show()
```

### Rename a column

```python
emp_df_rename = (
  emp_df.withColumnRenamed("name", "full_name")
)

emp_df_rename.show()
```

### Drop column

```python
emp_df_rename = emp_df_rename.drop("full_name")
```
