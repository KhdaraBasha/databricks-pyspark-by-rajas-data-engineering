# Databricks Utilities (dbutils) Functions

Databricks Utilities (`dbutils`) provide various utility commands for working with Databricks notebooks, file systems, widgets, and more.

## Available dbutils Modules

### 1. **dbutils.fs** - File System Utilities
Interact with cloud storage and DBFS:
* `dbutils.fs.ls(path)` - List files/directories
* `dbutils.fs.cp(from, to, recurse=False)` - Copy files
* `dbutils.fs.mv(from, to, recurse=False)` - Move files
* `dbutils.fs.rm(path, recurse=False)` - Remove files
* `dbutils.fs.mkdirs(path)` - Create directories
* `dbutils.fs.head(path, maxBytes=65536)` - Preview file content
* `dbutils.fs.put(path, content, overwrite=False)` - Write string to file
* `dbutils.fs.mount()` / `dbutils.fs.unmount()` - Mount cloud storage

### 2. **dbutils.notebook** - Notebook Workflow Utilities
Orchestrate notebook execution:
* `dbutils.notebook.run(path, timeout_seconds, arguments={})` - Run another notebook
* `dbutils.notebook.exit(value)` - Exit notebook with a value
* `dbutils.notebook.entry_point.getDbutils()` - Get dbutils instance

### 3. **dbutils.widgets** - Widget/Parameter Utilities
Create interactive parameters:
* `dbutils.widgets.text(name, defaultValue, label)` - Text input widget
* `dbutils.widgets.dropdown(name, defaultValue, choices, label)` - Dropdown widget
* `dbutils.widgets.combobox(name, defaultValue, choices, label)` - Combobox widget
* `dbutils.widgets.multiselect(name, defaultValue, choices, label)` - Multiselect widget
* `dbutils.widgets.get(name)` - Get widget value
* `dbutils.widgets.remove(name)` - Remove widget
* `dbutils.widgets.removeAll()` - Remove all widgets

### 4. **dbutils.secrets** - Secret Management
Access secrets from secret scopes:
* `dbutils.secrets.get(scope, key)` - Retrieve secret value
* `dbutils.secrets.list(scope)` - List secrets in scope
* `dbutils.secrets.listScopes()` - List available scopes

### 5. **dbutils.jobs** - Jobs Utilities
Access job context and metadata:
* `dbutils.jobs.taskValues.set(key, value)` - Set task value
* `dbutils.jobs.taskValues.get(taskKey, key, default, debugValue)` - Get task value

### 6. **dbutils.library** - Library Utilities
Manage libraries (legacy - prefer %pip):
* `dbutils.library.install(library)` - Install library
* `dbutils.library.installPyPI(package)` - Install PyPI package
* `dbutils.library.list()` - List installed libraries
* `dbutils.library.restartPython()` - Restart Python kernel

### 7. **dbutils.data** - Data Utilities
Manage data summaries:
* `dbutils.data.summarize(df)` - Generate data profile

## Common Usage Examples

```python
# List files in a directory
files = dbutils.fs.ls('/FileStore/')

# Get widget parameter
date_param = dbutils.widgets.get('date')

# Retrieve secret
api_key = dbutils.secrets.get(scope='my-scope', key='api-key')

# Run another notebook and get result
result = dbutils.notebook.run('./child-notebook', 60, {'param1': 'value1'})
```
